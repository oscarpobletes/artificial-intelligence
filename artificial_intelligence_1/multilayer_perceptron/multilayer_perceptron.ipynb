```python
# Libraries that we are going to use
import numpy as np # Linear algebra
```

```python
# Activation function: Sigmoid
def sigmoid (a): # a is the argument that will receive from the products that are made between weights and inputs
  return 1/(1+np.exp(-a)) 
```

```python
# Sigmoid implementation
def outSig(ai):
  if ai>=0.5:
    y=1
  else:
    y=0
  return y
```

```python
# Forward propagation 
def feedForward(w,input):
  length=len(input) # Get the input dimension
  y1= np.zeros((length,1)) # Fill the vectors with zeros
  y2= np.zeros((length,1))
  y=  np.zeros((length,1))

# Calculate the products
  for i in range (0,length):
    x=input[i,:] # Row i and all the columns 
    # Calculate the dot product for the layers
    # Input
    a1=np.dot(w[0],x) # Vector starts in zero
    a2=np.dot(w[1],x)
    # Hidden
    y1=sigmoid(a1)
    y2=sigmoid(a2)
    rsig=[y1,y2,1] # Save the results with bias to calculate a3
    a3=np.dot(w[2],rsig)
    # Ouput
    y1=sigmoid(a3) # Overwrite y1
    y[i]=outSig(y1) 
    # Verify
    #y[i]=y1
  return y
```
